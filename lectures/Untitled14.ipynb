```julia
function hilbertmatrix(m, n)
    ret = zeros(m, n)
    for j = 1:n, k=1:m
        ret[k,j] = 1/(k+j-1)
    end
    ret
end

hilbertmatrix(4,5)
```

```julia
scatter(svdvals(hilbertmatrix(1000,1001)); yscale=:log10)
```

```julia
H = hilbertmatrix(1000,1001)
U,σ,V = svd(H)
k = 15
H_k = U[:,1:k] * Diagonal(σ[1:k]) * V[:,1:k]'

opnorm(H_k - H)
```
