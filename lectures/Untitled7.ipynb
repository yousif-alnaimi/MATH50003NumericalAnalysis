```julia
using LinearAlgebra
n = 100
A = randn(n, n)
b = randn(n)
A[1,1] = 0.0001
A
```

```julia
Q,R = qr(A)
norm(A\b - R\(Q'b))
```

```julia
L,U = lu(A, Val(false))

norm((A \ b) - U\(L\b))
```

```julia
L,U,p = lu(A)
P = I(n)[p,:]
A ≈ P'*L*U

norm(A\b - U\(L\b[p]))
```

```julia
norm(big.(A)\big.(b) - U\(L\b[p]))
```

```julia
norm(big.(A)\big.(b) - R\(Q'b))
```

# LU Decomposition

```julia
L₁ = Matrix(1.0I, n, n)
L₁[2:end,1] = -A[2:end,1]/A[1,1]

A₁ = L₁*A

L₂ = Matrix(1.0I, n, n)
L₂[3:end,2] = -A₁[3:end,2]/A₁[2,2]


L₂*L₁*A

Ls = Vector{Matrix{Float64}}(undef, n-1)
Ã = copy(A)
for j = 1:n-1
    Ls[j] = Matrix(1.0I, n, n)
    Ls[j][j+1:end,j] = -Ã[j+1:end,j]/Ã[j,j]
    Ã = Ls[j]*Ã
end

U = UpperTriangular(Ã)


# U = Ls[n-1]…Ls[2]Ls[1]A ⇒ L = inv(Ls[1])…inv(Ls[n])

L = LowerTriangular(Matrix(1.0I, n, n))
for j = 1:n-1
    L[j+1:end,j] = -Ls[j][j+1:end,j]
end

L*U
```
